```python
import pandas as pd
```

```python
dataset = pd.read_csv('C:/Users/user/Desktop/DATA/sales_used untampered.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset['Date'] = pd.to_datetime(dataset['Date'])
```

```python
# TOTAL SALES BY CITY
city_sales = dataset.groupby('City')['Total'].sum().sort_values(ascending=False)
```

```python
city_sales.plot(kind='bar',title='Total Sales By City',figsize=(8,4),color='purple')
plt.ylabel('Sales Amount')
plt.tight_layout()
plt.show()
```

```python
# TOTAL SALES BY GENDER
gender_sales = dataset.groupby('Gender')['Total'].sum()
```

```python
gender_sales.plot(kind='bar',color='green',title='Total Sales by Gender')
plt.ylabel('Sales Amount')
plt.show()
```

```python
# DAILY SALES TREND
daily_sales = dataset.groupby('Date')['Total'].sum()
```

```python
daily_sales.plot(figsize=(10,5),title='Daily Sales Trend',color='Orange')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# TOP 5 HIGHEST TRANSACTIONS
top_sales = dataset.sort_values('Total',ascending=False).head(5)
print(top_sales[['Invoice ID','Date','City','Total']])
```
